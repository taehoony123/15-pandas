```python
import requests
import pandas as pd

my_key = ""

def get_geocoding(place):
    url = f"https://dapi.kakao.com/v2/local/search/address.json?query={place}"
    headers = {
        "Authorization": f"KakaoAK {my_key}"
    }
    response = requests.get(url,headers=headers)
    result = response.json()
    
    return result["documents"][0]["y"], result["documents"][0]["x"]
    
# get_geocoding("서울특별시 종로구 세종대로 175")

places = [
    "서울특별시 종로구 세종대로 175",
    "서울특별시 서초구 서초동 700",
    "부산광역시 해운대구 해운대해변로 264"
]
lat = []
lon = []

i = 0
for place in places:
    i = i + 1
    try:
        print(i,place)
        place_lat, place_lon = get_geocoding(place)
        lat.append(place_lat)
        lon.append(place_lon)
    except:
        lat.append("")
        lon.append("")
        print(i)

print(lat)
print(lon)

dict_data = {
    "위도" : lat,
    "경도" : lon
}

df = pd.DataFrame(dict_data)
df
```

## 국토교통부 아파트 매매 실거래가 자료

```python
# data.go.kr
## 국토교통부 아파트 매매 실거래가 자료
import requests
import xml.etree.ElementTree as ET
import pandas as pd

url = "https://apis.data.go.kr/1613000/RTMSDataSvcAptTrade/getRTMSDataSvcAptTrade"
SERVICE_KEY = ""

def fetch_apt_trade(serviceKey,LAWD_CD,DEAL_YMD):

    params = {
        "serviceKey" : serviceKey,
        "LAWD_CD" : LAWD_CD,
        "DEAL_YMD" : DEAL_YMD
    }

    response = requests.get(url, params = params)
    # print(response.content)

    root = ET.fromstring(response.content)
    # print(root)

    items = root.findall(".//item")
    # print(items)

    records = []
    for item in items:
        row = {}

        for child in item:
            row[child.tag] = child.text

        records.append(row)

    return records


data = fetch_apt_trade(SERVICE_KEY , LAWD_CD="27230" , DEAL_YMD="202604")
df = pd.DataFrame(data)
df.head()
```
