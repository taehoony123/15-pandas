## 행 인덱스 초기화 .reset_index

```python

import pandas as pd

dict_data = {
    "c0" : [1,2,3],
    "c1" : [4,5,6],
    "c2" : [7,8,9],
    "c3" : [10,11,12],
    "c4" : [13,14,15]
}

df = pd.DataFrame(dict_data, index=['r0','r1','r2'])
print(df)
```

```python
ndf = df.reset_index(names = ['C00'])
print(ndf)
```

```python
ndf2 = df.reset_index(drop=True)
print(ndf2)
```

## 행 인덱스 기준으로 df 정렬 .sort_index(ascending= True,False)

```python
import pandas as pd

dict_data = {
    "c0" : [1,2,3],
    "c1" : [4,5,6],
    "c2" : [7,8,9],
    "c3" : [10,11,12],
    "c4" : [13,14,15]
}

df = pd.DataFrame(dict_data, index=['r0','r2','r1'])
print(df)

ndf3 = df.sort_index(ascending=True)
print(ndf3)
```

```python
import pandas as pd

dict_data = {
    "c0" : [1,2,3],
    "c1" : [4,5,6],
    "c2" : [7,8,9],
    "c3" : [10,11,12],
    "c4" : [13,14,15]
}

df = pd.DataFrame(dict_data, index=['r0','r1','r2'])

nff2 = df.sort_index(ascending=True)
print(nff2)
ndf = df.sort_values(by='c4',ascending=False)
ndf
```
