```python
## 시리즈 vs 시리즈
import pandas as pd

student1 = pd.Series({
    '국어': 100,
    '영어' : 80,
    '수학' : 90
})

student2 = pd.Series({
    '수학' : 80,
    '국어' : 90,
    '영어' : 80
})

print(student1)
print()
print(student2)

addition = student1 +student2
print()
print(type(addition))
print(addition)
```

```python
subtraction = student1 - student2
print(subtraction)
```

```python
multiplication = student1 * student2
division = student1 / student2

print(multiplication)
print(division)
```

```python
result = pd.DataFrame([addition, subtraction, multiplication, division], index = ['덧셈', '뺄셈', '곱셈', '나눗셈'])
result
```

```python
## nan

import pandas as pd
import numpy as np

student1 = pd.Series({
    '국어': np.nan,
    '영어' : 80,
    '수학' : 90
})

student2 = pd.Series({
    '수학' : 80,
    '국어' : 90
})

print(student1)
print()
print(student2)

addition = student1 +student2
subtraction = student1 - student2
multiplication = student1 * student2
division = student1 / student2
print()

result = pd.DataFrame([addition, subtraction, multiplication, division], index = ['덧셈', '뺄셈', '곱셈', '나눗셈'])
result
```

```python
## nan

import pandas as pd
import numpy as np

student1 = pd.Series({
    '국어': np.nan,
    '영어' : 80,
    '수학' : 90
})

student2 = pd.Series({
    '수학' : 80,
    '국어' : 90
})

print(student1)
print()
print(student2)

sr_add = student1.add(student2, fill_value=0)
sr_sub = student1.sub(student2, fill_value=0)
sr_mul = student1.mul(student2, fill_value=0)
sr_div = student1.div(student2, fill_value=0)
print()

result = pd.DataFrame([sr_add, sr_sub, sr_mul, sr_div], index = ['덧셈', '뺄셈', '곱셈', '나눗셈'])
result
```

## df vs df

```python
import pandas as pd
import seaborn as sns

titanic = sns.load_dataset('titanic')
df = titanic.loc[:,['age','fare']]
print(df.tail())
print()
addition = df + 10
print(addition.tail())
print()

sub = addition - df
print(sub.tail())
```

```python
sample1 = addition.tail()
sample2 = sub.tail().fillna(0.0)

print(sample1)
print(sample2)
```

```python
df_add = sample1.add(sample2,fill_value=0)
df_sub = sample1.sub(sample2,fill_value=0)
df_mul = sample1.mul(sample2,fill_value=0)
df_div = sample1.div(sample2,fill_value=0)

print(df_add)
print(df_sub)
print(df_mul)
print(df_div)
```

## filtering

```python
import pandas as pd
import seaborn as sns

titanic = sns.load_dataset('titanic')

df = titanic.loc[0:9,['age','fare']]
df
```

```python
df['age'] < 20
```

```python
df[df['age'] < 20]
```

```python
mask1 = (titanic.age >= 10) & (titanic.age < 20)
df_teenage = titanic[mask1]
df_teenage.head()
```

```python
mask2 = (titanic.age < 10 )& (titanic.sex == "female")
df_female_under10 = titanic[mask2]
df_female_under10.head()
```

```python
mask3 = (titanic.age < 10 ) | (titanic.age >= 60)
df_under10_morethan60 = titanic.loc[mask3,['age','sex','alone']]
df_under10_morethan60.head()
```
