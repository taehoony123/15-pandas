## 문자열 메서드

```python
import pandas as pd
import numpy as np

ser = pd.Series(["Apple_사과","Banana_바나나","Cherry_체리",np.nan],index = ["First ", " Second", " Third", "Fourth"])

print(ser)
print()

ser2 = pd.Series(["Apple_사과","Banana_바나나","Cherry_체리",np.nan],index = ["First ", " Second", " Third", "Fourth"], dtype='string')
print(ser2)
```

```python
print(ser.str.lower())
print()

print(ser.str.upper())
print()

print(ser.str.len())
```

```python
# 문자열 반환
print(ser.str.split("_"))
print()

# 문자열을 분할하여 DataFrame으로 확장
print(ser.str.split("_",expand=True))
print()

# 분할된 문자열 중 특정 요소 선택(1)
print(ser.str.split("_").str.get(1))
```

```python
# 행 인덱스 추출
idx = ser.index
print(idx)

# 양쪽 공백 제거
print(idx.str.strip())

# 왼쪽 공백 제거
print(idx.str.lstrip())

# 오른쪽 공백 제거
print(idx.str.rstrip())
```

```python
# 문자열 교체
print(ser.str.replace("_", ":", regex=False)) #regex=False는 정규표현식을 사용하지 않고, 단순한 문자열을 찾아서 변경하는 것
print()

# 정규 표현식을 사용한 문자열 교체 (영문 알파벳이 아닌 문자 제거)
print(ser.str.replace(r"[^a-zA-Z\s]","",regex=True))
print()
print(ser.str.replace("[^a-zA-Z\\s]","",regex=True))
```

```python
print(ser.str[0])
print()

print(ser.str[0:4])
```

문자열 데이터에서 특정 문자나 패턴의 포함 여부를 확인하고 전체 문자열이

특정 패턴과 일치하는지 여부를 검사하는 방법

```python
print(ser)
```

```python
# 특정 문자 포함 여부 확인, NA값을 False로 처리
contains_A = ser.str.contains("A",na=False)
print(contains_A)
print()

# 패턴 포함 여부 확인
contains_pattern = ser.str.contains(r"[A|B][a-z]+")
print(contains_pattern)
print()

# 전체 문자열의 패턴 일치 여부 확인
fullmatch_pattern = ser.str.fullmatch(r"[A|B][a-z]+")
# 뒤에 한글이나 _ 포함되어 있기 떄문에 문자열 전체 패턴과 일치 안함
print(fullmatch_pattern)
```

```python
# 새로운 시리즈 생성
ser3 = pd.Series(["a1","b2","c3","d4a5","e6e7","f8"],dtype="string")

# 숫자만 추출
digits = ser3.str.extractall(r"(\d)").groupby(level=0).agg(''.join)
print(digits)
print()
# 알파벳만 추출
letters = ser3.str.extractall(r"([a-z])").groupby(level=0).agg(''.join)
print(letters)
```

```python
split_columns = ser3.str.extract(r"(?P<Letter>[a-z]+)(?P<Digit>\d+)")
```
